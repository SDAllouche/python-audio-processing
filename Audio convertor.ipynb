```python
import tkinter
from tkinter import *
from tkinter import messagebox
from tkinter.filedialog import *
import json
from tkinter import ttk
from tkinter.colorchooser import *
import wave
import pydub
from pydub import AudioSegment
import os
import ffmpeg
import matplotlib.pyplot as plt
import numpy
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import soundfile as sf
import pygame
```

```python
window =Tk()
window.title("Audio Converter")
window.geometry("1200x480")
window.resizable(width=False,height=False)
window.config(bg="#E8E8E8")

def open_file ():
    global file
    file = filedialog.askopenfilename()
    e1['state']='normal' ; e1.delete(0,END) ; e1.insert(0,file)
    btn_nc['state'] = 'normal'
    btn_ply_ori['state'] = 'normal'; btn_stp_ori['state'] = 'normal'
    conditions(file)
    if str(file).endswith(".wav")==True or str(file).endswith(".flac") or str(file).endswith(".ogg"):
        btn_plt_ori['state'] = 'normal'
    else:
        btn_plt_ori['state'] = 'disable'

def channels ():
    aud = AudioSegment.from_file(file=file)
    nc= aud.channels
    e3['state'] = 'normal'
    if(nc==1):
        e3.delete(0, END)
        e3.insert(0,"Mono")
    elif(nc==2):
        e3.delete(0, END)
        e3.insert(0,"Stereo")
        c.config(state='normal')

def ext(path):
    data,fr= sf.read(path)
    return data

def play1(path):
    pygame.init()
    pygame.mixer.music.load(path)
    pygame.mixer.music.play(loops=0)

def stop():
    pygame.init()
    pygame.mixer.music.stop()

def split(path):
    data_hex=ext(path)
    vect1=[data_hex[i] for i in range(len(data_hex)) if(i%2==0)]
    vect2=[data_hex[i] for i in range(len(data_hex)) if(i%2!=0)]
    vect=[vect1,vect2]
    return vect

def graph(path):
    aud = AudioSegment.from_file(file=path)
    ncl = aud.channels
    n=v.get()
    if n==1:
        top = Toplevel()
        top.geometry("1000x215")
        vect=ext(path)
        a=plt.figure(figsize=(20,3))
        a.add_subplot(111).plot(vect)
        canvas1 = FigureCanvasTkAgg(a,top)
        canvas1.get_tk_widget().pack(side=TOP)
    else :
        if ncl==1:
            top = Toplevel()
            top.geometry("1000x215")
            vect=ext(path)
            a=plt.figure(figsize=(20,3))
            a.add_subplot(111).plot(vect)
            canvas1 = FigureCanvasTkAgg(a,top)
            canvas1.get_tk_widget().pack(side=TOP)
        elif ncl==2:
            top = Toplevel()
            top.geometry("1000x430")
            vect=split(path)
            a1=plt.figure(figsize=(20,3))
            a1.add_subplot(111).plot(vect[0])
            canvas1 = FigureCanvasTkAgg(a1,top)
            canvas1.get_tk_widget().pack(side=TOP)
            a2 = plt.figure(figsize=(20,3))
            a2.add_subplot(111).plot(vect[1])
            canvas2 = FigureCanvasTkAgg(a2,top)
            canvas2.get_tk_widget().pack(side=TOP)
            

def browser():
    global compresse,form,bit
    directory= askdirectory()
    name=e4.get()
    form=click.get()
    bit=click2.get()
    if form=="mp3":
        chemin=directory+"/"+name+"_"+bit+"."+form
    else:
        chemin=directory+"/"+name+"."+form
    e2['state'] = 'normal' ; e2.delete(0, END) ; e2.insert(0,chemin)
    compresse=e2.get()
    btn_comp.config(state='normal')
    
def conditions (file):
    taill = len(str(file))
    liste = [i for i in str(file[taill - 4:taill]) if i != "."]
    formate = "".join(liste)
    e4['state'] = 'normal' ; e4.delete(0, END)
    btn_browser['state'] = 'normal'
    drop.config(state='normal') ; drop['menu'].entryconfigure(formate,state='disable')
    drop2.config(state='normal')

def compress(path,path2,formate,bit):
    audio= AudioSegment.from_file(path)
    if formate=='mp3':
        audio.export(path2,format=formate,bitrate=bit)
    else:
        audio.export(path2,format=formate)
    resp = messagebox.showinfo("Convert","You convetr audio successfully")
    btn_plt_comp.config(state='normal');btn_ply_comp.config(state='normal');btn_stp_comp.config(state='normal')
    
ply=PhotoImage('./Icons/play50.png')
stp=PhotoImage('./Icons/stop-button50.png')


frame1=LabelFrame(window,bg="#E8E8E8",text="Original  :",font=("",12))
frame1.grid(row=0,column=0,padx=2,pady=20,ipadx=0,ipady=100)

frame2=LabelFrame(window,bg="#E8E8E8",text="Convert  :",font=("",12))
frame2.grid(row=0,column=1,padx=0,pady=20,ipadx=8,ipady=20)


btn_ply_ori=Button(frame1,borderwidth=0,command=lambda : play1(file),state='disable',bg="#E8E8E8")
btn_stp_ori=Button(frame1,borderwidth=0,command=stop,state='disable',bg="#E8E8E8")
btn_get=Button(frame1,text='Upload Audio',command=open_file)
e1=Entry(frame1,width=70,state='disable')
mylabel = Label(window,)
btn_plt_ori=Button(frame1,text="Graph",state='disable',command=lambda :graph(file))
btn_nc=Button(frame1,text='Pistes',command=channels,state='disable')
v = IntVar()
c = Checkbutton(frame1, text="Only one plot", variable=v,state='disable')
c.deselect()
e3=Entry(frame1,state='disable')

e1.grid(row=0,column=0,padx=10,pady=20,ipady=5,ipadx=6)
btn_get.grid(row=0,column=1,padx=10,ipadx=10,ipady=3,columnspan=2)
btn_nc.grid(sticky=W,row=1,column=0,padx=10,pady=0,columnspan=2,ipadx=20,ipady=5)
e3.grid(sticky=W,row=1,column=0,padx=110,pady=0,columnspan=2,ipadx=50,ipady=5)
btn_ply_ori.grid(sticky=E,row=1,column=0,padx=80,pady=0,columnspan=2)
btn_stp_ori.grid(sticky=E,row=1,column=0,padx=0,pady=0,columnspan=2)
btn_plt_ori.grid(sticky =W,row=2,column=0,padx=(200,0),pady=20,ipadx=100,ipady=5,columnspan=2)
c.grid(sticky =W,row=2,column=0,columnspan=2,padx=(40,0),pady=20)


btn_ply_comp=Button(frame2,borderwidth=0,state='disable',command=lambda:play1(compresse),bg="#E8E8E8")
btn_stp_comp=Button(frame2,borderwidth=0,state='disable',command=stop,bg="#E8E8E8")
btn_comp=Button(frame2,text='Start Converting ',state='disable',command=lambda :compress(file,compresse,form,bit))
e2=Entry(frame2,width=70,state='disable')
e4=Entry(frame2,width=30,state='disable')
btn_browser=Button(frame2,text="Browser...",state='disable',command=browser)
mylabel1 = Label(frame2, text="Chose the format : ")
mylabel2 = Label(frame2, text="Bitrate : ")
mylabel3 = Label(frame2, text="New name file : ")
formate = ["wav","flac","mp3","ogg","m4a"]
click = StringVar(window)
click.set("Format")
drop = OptionMenu(frame2, click, *formate)
drop.config(state='disable')
bit = ["16k","64k","128k","256k","320k"]
click2 = StringVar()
click2.set("Non")
drop2 = OptionMenu(frame2, click2, *bit)
drop2.config(state='disable')

mylabel3.grid(sticky =W,row=0,column=0,padx=20,pady=(20,30))
e4.grid(sticky =W,row=0,column=0,padx=(160,0),pady=(20,30),ipadx=10,ipady=5)
btn_ply_comp.grid(sticky =W,row=4,column=0,padx=(180,0),pady=30)
btn_stp_comp.grid(sticky =W,row=4,column=0,padx=(310,0),pady=30)
btn_comp.grid(sticky =W+E,row=3,column=0,padx=80,ipadx=40,ipady=5,columnspan=2)
drop.grid(sticky =W,row=1,column=0,padx=150,ipadx=10,ipady=3)
e2.grid(sticky =W,row=2,column=0,padx=15,pady=30,ipadx=10,ipady=5,columnspan=2)
btn_browser.grid(row=2,column=0,padx=(486,0),pady=30,ipadx=20,ipady=3,columnspan=2)
mylabel1.grid(sticky =W,row=1,column=0,padx=10)
drop2.grid(sticky =E,row=1,column=0,padx=(0,0),ipadx=10,ipady=3)
mylabel2.grid(sticky =E,row=1,column=0,padx=(0,240),columnspan=2)


window.mainloop()
```
